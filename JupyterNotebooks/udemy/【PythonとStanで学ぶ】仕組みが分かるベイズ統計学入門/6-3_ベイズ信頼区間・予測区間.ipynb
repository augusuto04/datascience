```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pystan
from scipy.stats import mstats
%matplotlib inline
```

```python
plt.style.use("ggplot")
```

```python
df = pd.read_excel("./data/real_estate1.xlsx")
```

```python
stan_model = """
data {
    int N;
    real X[N];
    real Y[N];
    int N_s;
    real X_s[N_s];
}
parameters {
    real a;
    real b;
    real<lower=0> sigma;
}
model {
    for (n in 1:N) {
        Y[n] ~ normal(a * X[n] + b, sigma);
    }
}
generated quantities {
    real Y_s[N_s];
    for (n in 1:N_s) {
        Y_s[n] = normal_rng(a * X_s[n] + b, sigma);
    }
}
"""
```

```python
sm = pystan.StanModel(model_code=stan_model)
```

```python
X_s = np.arange(40, 90, 1)
N_s = X_s.shape[0]
stan_data = {"N": df.shape[0], "X": df["space"], "Y": df["value"], "N_s": N_s, "X_s": X_s}
```

```python
fit = sm.sampling(data=stan_data, iter=2000, warmup=500, chains=3, seed=123)
```

```python
fit.extract("a")
```

```python
ms_a = fit.extract("a")["a"]
```

```python
plt.hist(ms_a)
```

```python
ms_b = fit.extract("b")["b"]
```

```python
df_b = pd.DataFrame([])
for i in range(40, 90, 1):
    df_b[i] = ms_a * i + ms_b
```

```python
df_b
```

```python
low_y50, high_y50 = mstats.mquantiles(df_b, [0.25, 0.75], axis=0)
low_y90, high_y90 = mstats.mquantiles(df_b, [0.05, 0.95], axis=0)
```

```python
plt.scatter(df["space"], df["value"])
plt.fill_between(X_s, low_y50, high_y50, alpha=0.6, color="darkgray")
plt.fill_between(X_s, low_y90, high_y90, alpha=0.3, color="gray")
a = 77.99
b = -692.9
y = a * X_s + b
plt.plot(X_s, y, color="black")
```
